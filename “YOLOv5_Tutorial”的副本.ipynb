Validate a model's accuracy on the [COCO](https://cocodataset.org/#home) dataset's `val` or `test` splits. Models are downloaded automatically from the [latest YOLOv5 release](https://github.com/ultralytics/yolov5/releases). To show results by class use the `--verbose` flag.

````
# Âä†ËΩΩ API keyÔºà‰ªé Colab SecretsÔºåÂøÖÈ°ªÈáçÂêØ Runtime ÂêéËøêË°åÔºâ
from google.colab import userdata
API_KEY = userdata.get('ROBOFLOW_API_KEY')
print("‚úì API key Âä†ËΩΩÊàêÂäüÔºÅ")

!pip install roboflow tqdm -q

from roboflow import Roboflow
import os, shutil
from tqdm import tqdm

rf = Roboflow(api_key=API_KEY)

# ÊúÄÁªàËæìÂá∫Ë∑ØÂæÑ
final_path = "/content/medical_merged_15class"
os.makedirs(final_path, exist_ok=True)

# Áªü‰∏Ä 15 ‰∏™Á±ªÂà´
classes = ["Scalpel","Forceps","Scissors","Needle Holder","Retractor",
           "Clamp","Suture","Tweezers","Cautery","Suction Tube",   # 0~9 ÊâãÊúØÂ∑•ÂÖ∑
           "Gown","Mask","Gloves","Goggles","Person"]             # 10~14 Èò≤Êä§Ë£ÖÂ§á

print("ÂºÄÂßã‰∏ãËΩΩ‰∏§‰∏™Êï∞ÊçÆÈõÜÂπ∂ÂêàÂπ∂ÔºàÁ∫¶ 1~2 ÂàÜÈíüÔºâ...")

# 1. ‰∏ãËΩΩÊâãÊúØÂ∑•ÂÖ∑Êï∞ÊçÆÈõÜÔºà‰øÆÊ≠£ workspace: "surgical-tools"Ôºåv1Ôºå972 Âº†ÂõæÔºå10+ Á±ªÔºâ
print("\n1. ‰∏ãËΩΩÊâãÊúØÂ∑•ÂÖ∑Êï∞ÊçÆÈõÜ...")
surg_project = rf.workspace("surgical-tools").project("surgical-tools-dataset")
surg_dataset = surg_project.version(1).download("yolov8", location="./surg_temp")

# 2. ‰∏ãËΩΩÂåªÊä§Èò≤Êä§Ë£ÖÂ§áÊï∞ÊçÆÈõÜÔºàv7Ôºå11978 Âº†ÂõæÔºå5 Á±ªÔºâ
print("\n2. ‰∏ãËΩΩÂåªÊä§Èò≤Êä§Ë£ÖÂ§áÊï∞ÊçÆÈõÜ...")
ppe_project = rf.workspace("personal-protective-equipment").project("ppes-kaxsi")
ppe_dataset = ppe_project.version(7).download("yolov8", location="./ppe_temp")

# ÂêàÂπ∂ÂáΩÊï∞ÔºàÂ§çÂà∂ + ID ÂÅèÁßªÔºâ
def merge(src_dir, prefix, offset=0):
    for split in ["train", "val", "test"]:
        img_src = f"{src_dir}/{split}/images"
        lbl_src = f"{src_dir}/{split}/labels"
        if not os.path.exists(img_src):
            print(f"  Ë∑≥Ëøá {split}ÔºàÊó†Êï∞ÊçÆÔºâ")
            continue

        img_dst = f"{final_path}/{split}/images"
        lbl_dst = f"{final_path}/{split}/labels"
        os.makedirs(img_dst, exist_ok=True)
        os.makedirs(lbl_dst, exist_ok=True)

        print(f"  Â§ÑÁêÜ {prefix} {split}")
        files = [f for f in os.listdir(img_src) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]
        for img in tqdm(files):
            new_name = prefix + img
            new_lbl = prefix + os.path.splitext(img)[0] + ".txt"

            # Â§çÂà∂ÂõæÁâá
            shutil.copy2(f"{img_src}/{img}", f"{img_dst}/{new_name}")

            # Â§ÑÁêÜÊ†áÁ≠æ
            old_lbl = os.path.splitext(img)[0] + ".txt"
            lbl_path = f"{lbl_src}/{old_lbl}"
            if os.path.exists(lbl_path):
                with open(lbl_path) as f:
                    lines = f.readlines()
                if offset > 0:
                    new_lines = []
                    for line in lines:
                        if line.strip():
                            parts = line.strip().split()
                            old_id = int(parts[0])
                            # Êò†Â∞Ñ PPE ID: ÂÅáËÆæÂéüÂßã 0=Gown‚Üí10, 1=Mask‚Üí11 Á≠â
                            new_id = old_id + offset
                            new_line = f"{new_id} {' '.join(parts[1:])}\n"
                            new_lines.append(new_line)
                        else:
                            new_lines.append(line)
                    lines = new_lines
                with open(f"{lbl_dst}/{new_lbl}", "w") as f:
                    f.writelines(lines)
            else:
                # Á©∫Ê†áÁ≠æ
                with open(f"{lbl_dst}/{new_lbl}", "w") as f:
                    pass

# ÊâßË°åÂêàÂπ∂
merge(surg_dataset.location, "surg_")        # ID 0~9
merge(ppe_dataset.location, "ppe_", 10)      # ID 10~14

# ÁîüÊàê data.yaml
yaml = f"""path: {final_path}
train: train/images
val: val/images
test: test/images

nc: 15
names: {classes}
"""
with open(f"{final_path}/data.yaml", "w") as f:
    f.write(yaml)

print("\n‚úì Êï∞ÊçÆÈõÜÂêàÂπ∂ÂÆåÊàêÔºÅ")
for s in ["train","val","test"]:
    img_dir = f"{final_path}/{s}/images"
    lbl_dir = f"{final_path}/{s}/labels"
    if os.path.exists(img_dir):
        n_img = len([f for f in os.listdir(img_dir) if f.lower().endswith(('.jpg', '.png', '.jpeg'))])
        n_lbl = len([f for f in os.listdir(lbl_dir) if f.endswith('.txt')])
        print(f"{s:5}: {n_img} Âº†ÂõæÁâáÔºå{n_lbl} ‰∏™Ê†áÁ≠æÊñá‰ª∂")

# Ê∏ÖÁêÜ
!rm -rf ./surg_temp ./ppe_temp

print(f"\nÊï∞ÊçÆÈõÜË∑ØÂæÑÔºö{final_path}")
````

````
import os
import shutil
import random
from tqdm import tqdm
import yaml

# ÂéüÂßãË∑ØÂæÑÔºà‰Ω†Áé∞Âú®Â∞±Ëøô‰∫õÔºâ
raw_path = "/content/medical_merged_15class"
final_path = "/content/medical_yolov5_final"

# Âà†Èô§ÊóßÁöÑÔºåÈáçÊñ∞Âª∫‰∏Ä‰∏™Âπ≤ÂáÄÁöÑ
if os.path.exists(final_path):
    shutil.rmtree(final_path)
os.makedirs(final_path, exist_ok=True)

# ÁõÆÊ†áÁªìÊûÑ
for split in ["train", "valid", "test"]:
    os.makedirs(f"{final_path}/{split}/images", exist_ok=True)
    os.makedirs(f"{final_path}/{split}/labels", exist_ok=True)

print("ÂºÄÂßãÈáçÊñ∞Êï¥ÁêÜÊï∞ÊçÆÈõÜÔºàËá™Âä®ÂàáÂàÜ val ÈõÜÔºâ...")

# 1. ÂÖàÊää train Âíå test ÁöÑÂÖ®ÈÉ®Â§çÂà∂ËøõÊù•
def copy_all(src_split, dst_split):
    img_src = f"{raw_path}/{src_split}/images"
    lbl_src = f"{raw_path}/{src_split}/labels"
    img_dst = f"{final_path}/{dst_split}/images"
    lbl_dst = f"{final_path}/{dst_split}/labels"

    if not os.path.exists(img_src):
        print(f"Ë∑≥Ëøá {src_split}Ôºà‰∏çÂ≠òÂú®Ôºâ")
        return 0

    count = 0
    for img in tqdm(os.listdir(img_src), desc=f"Â§çÂà∂ {src_split}"):
        if img.lower().endswith(('.jpg', '.jpeg', '.png')):
            shutil.copy2(f"{img_src}/{img}", f"{img_dst}/{img}")
            lbl = os.path.splitext(img)[0] + ".txt"
            if os.path.exists(f"{lbl_src}/{lbl}"):
                shutil.copy2(f"{lbl_src}/{lbl}", f"{lbl_dst}/{lbl}")
            count += 1
    return count

train_cnt = copy_all("train", "train")
test_cnt  = copy_all("test",  "test")
print(f"train: {train_cnt} Âº†, test: {test_cnt} Âº†")

# 2. ‰ªé train ‰∏≠ÈöèÊú∫Âàá 15% ‰Ωú‰∏∫ validÔºàÊ†áÂáÜÂÅöÊ≥ïÔºâ
all_train_imgs = os.listdir(f"{final_path}/train/images")
val_ratio = 0.15
val_count = int(len(all_train_imgs) * val_ratio)

random.seed(42)
val_imgs = random.sample(all_train_imgs, val_count)
print(f"‰ªé train ‰∏≠ÂàáÂá∫ {val_count} Âº†‰Ωú‰∏∫ valid...")

for img in tqdm(val_imgs):
    # ÁßªÂä®ÂõæÁâá
    shutil.move(f"{final_path}/train/images/{img}", f"{final_path}/valid/images/{img}")
    # ÁßªÂä®Ê†áÁ≠æ
    lbl = os.path.splitext(img)[0] + ".txt"
    if os.path.exists(f"{final_path}/train/labels/{lbl}"):
        shutil.move(f"{final_path}/train/labels/{lbl}", f"{final_path}/valid/labels/{lbl}")

# 3. ÂÜôÂÖ• YOLOv5 ÂÆåÁæé data.yamlÔºàÁõ∏ÂØπË∑ØÂæÑÁâàÔºâ
yaml_content = """train: train/images
val: valid/images
test: test/images

nc: 15
names: ['Scalpel','Forceps','Scissors','Needle Holder','Retractor',
        'Clamp','Suture','Tweezers','Cautery','Suction Tube',
        'Gown','Mask','Gloves','Goggles','Person']
"""

with open(f"{final_path}/data.yaml", "w") as f:
    f.write(yaml_content)

print("\nÊï∞ÊçÆÈõÜÊï¥ÁêÜÂÆåÊàêÔºÅ")
!ls -la {final_path}
print(f"\nÊúÄÁªàÁªìÊûÑÔºö")
!find {final_path} -type d | head -10

# ÁªüËÆ°ÊúÄÁªàÊï∞Èáè
for split in ["train", "valid", "test"]:
    imgs = len([f for f in os.listdir(f"{final_path}/{split}/images") if f.lower().endswith(('.jpg','.png','.jpeg'))])
    lbls = len([f for f in os.listdir(f"{final_path}/{split}/labels") if f.endswith('.txt')])
    print(f"{split:6}: {imgs} Âº†ÂõæÁâáÔºå{lbls} ‰∏™Ê†áÁ≠æ")

print(f"\nÂÆåÁæé YOLOv5 Êï∞ÊçÆÈõÜÂ∑≤Â∞±Áª™ÔºÅË∑ØÂæÑÔºö{final_path}")
````

````
# Ë∂ÖÁ∫ßÁ≤óÊö¥‰ΩÜ100%ÊúâÊïàÁöÑÁªàÊûÅ‰øÆÂ§çÔºàÊääÊâÄÊúâ >14 ÁöÑÁ±ªÂà´ÂÖ®ÈÉ®Âà†ÊéâÔºâ
import os
from tqdm import tqdm

def nuclear_fix(root):
    fixed = 0
    for split in ["train", "valid", "test"]:
        path = f"/content/medical_yolov5_final/{split}/labels"
        if not os.path.exists(path):
            continue
        print(f"Ê†∏ÂºπÁ∫ßÊ∏ÖÁêÜ {split} ...")
        for f in tqdm(os.listdir(path)):
            if not f.endswith(".txt"):
                continue
            fp = os.path.join(path, f)
            lines = []
            changed = False
            with open(fp) as file:
                for line in file:
                    if not line.strip():
                        lines.append(line)
                        continue
                    cls = int(line.split()[0])
                    if cls > 14:  # Ë∂ÖËøá14ÁöÑÂÖ®ÈÉ®Âà†Êéâ
                        changed = True
                        continue
                    lines.append(line)
            if changed:
                with open(fp, "w") as file:
                    file.writelines(lines)
                fixed += 1
    print(f"Ê†∏ÂºπÊ∏ÖÁêÜÂÆåÊàêÔºÅÂÖ±Ê∏ÖÁêÜ‰∫Ü {fixed} ‰∏™Êñá‰ª∂")

nuclear_fix("/content/medical_yolov5_final")
````

Close the active learning loop by sampling images from your inference conditions with the `roboflow` pip package
<br><br>

Train a YOLOv5s model on the [COCO128](https://www.kaggle.com/datasets/ultralytics/coco128) dataset with `--data coco128.yaml`, starting from pretrained `--weights yolov5s.pt`, or from randomly initialized `--weights '' --cfg yolov5s.yaml`.

- **Pretrained [Models](https://github.com/ultralytics/yolov5/tree/master/models)** are downloaded
automatically from the [latest YOLOv5 release](https://github.com/ultralytics/yolov5/releases)
- **[Datasets](https://github.com/ultralytics/yolov5/tree/master/data)** available for autodownload include: [COCO](https://github.com/ultralytics/yolov5/blob/master/data/coco.yaml), [COCO128](https://github.com/ultralytics/yolov5/blob/master/data/coco128.yaml), [VOC](https://github.com/ultralytics/yolov5/blob/master/data/VOC.yaml), [Argoverse](https://github.com/ultralytics/yolov5/blob/master/data/Argoverse.yaml), [VisDrone](https://github.com/ultralytics/yolov5/blob/master/data/VisDrone.yaml), [GlobalWheat](https://github.com/ultralytics/yolov5/blob/master/data/GlobalWheat2020.yaml), [xView](https://github.com/ultralytics/yolov5/blob/master/data/xView.yaml), [Objects365](https://github.com/ultralytics/yolov5/blob/master/data/Objects365.yaml), [SKU-110K](https://github.com/ultralytics/yolov5/blob/master/data/SKU-110K.yaml).
- **Training Results** are saved to `runs/train/` with incrementing run directories, i.e. `runs/train/exp2`, `runs/train/exp3` etc.
<br>

A **Mosaic Dataloader** is used for training which combines 4 images into 1 mosaic.

## Local Logging

Training results are automatically logged with [Tensorboard](https://www.tensorflow.org/tensorboard) and [CSV](https://github.com/ultralytics/yolov5/pull/4148) loggers to `runs/train`, with a new experiment directory created for each new training as `runs/train/exp2`, `runs/train/exp3`, etc.

This directory contains train and val statistics, mosaics, labels, predictions and augmentated mosaics, as well as metrics and charts including precision-recall (PR) curves and confusion matrices.
